```python
# a. Convert all the â€˜?â€™ values to NaN. Clean the data as per further requirements.
df.replace('?',np.nan,inplace = True)
df
```

```python
df.rename(columns= {'normalized-losses':'loss'},inplace = True)
```

```python
df.rename(columns= {'make':'company'} ,inplace = True)
```

```python
# b Which company manufactured the most expensive car and at what price?
df1 = df.sort_values(by = ['price'],ascending = False)
x12 = df1[['company', 'price']][0:5]
x12
```

```python
# c. Calculate the maximum horsepower for each company.
car_Manufacturers = df.groupby('company')
priceDf = car_Manufacturers[['company','horsepower']].max()
priceDf['horsepower']
```

```python
import seaborn as sns
sns.set(rc = {'figure.figsize':(10,5)})
sns.countplot(x=priceDf.horsepower)
plt.show()
```
