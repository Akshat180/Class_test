### Q. 8) In how many movies/shows, Tom Cruise was cast ?

```python

df2 = df.copy() 
df2.dropna(inplace = True)
df2['New'] = df2['cast'].apply(lambda a : str(a.split(",")) if a else np.nan)
df2['New'] = df2['New'].apply(lambda x: 'Tom Cruise' if 'Tom Cruise' in x else x)
df2[df2['New']=='Tom Cruise']
```
